```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df= pd.read_excel('PostCorrel.xlsx')
df.head()
```

```python
Y=df['WonChampionship']
X=df[['WasFirstSeed', 'WinPercent', 'PTS', 'FGPercent', '3PPercent', 'FTPercent', 'REB', 'TOV', 'BLK', 'PF', 'FTARank', 'PTSRank', '3PARank']]
X=sm.add_constant(X)
```

```python
log_reg = sm.Logit(Y,X).fit()
log_reg.summary()
```
